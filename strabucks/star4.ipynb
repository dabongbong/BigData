```python
import pandas as pd
import folium
import json
```

```python
seoul_starbucks = pd.read_excel('seoul_starbucks_list.xlsx')
seoul_starbucks
```

```python
starbucks_map = folium.Map(
    location=[37.57, 126.98],
    zoom_start=12,
    tiles='Stamen Terrain'


)
starbucks_map
```

```python
for idx in seoul_starbucks.index:
    lat = seoul_starbucks.loc[idx,'위도']
    lng = seoul_starbucks.loc[idx,'경도']

    folium.CircleMarker(
        location=[lat,lng],
        fill = True, 
        fill_color = 'green', 
        fill_opacity = 1, 
        color = 'yellow', 
        weigth = 1, 
        radius = 3,
    ).add_to(starbucks_map)

starbucks_map
```
