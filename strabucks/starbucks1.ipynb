```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome('./chromedriver')
driver.get('https://www.starbucks.co.kr/store/store_map.do')
```

```python
find = '#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a'
driver.find_element_by_css_selector(find).click()
```

```python
seoul = '#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a'
driver.find_element_by_css_selector(seoul).click()
```

```python
all = '#mCSB_2_container > ul > li:nth-child(1) > a'
driver.find_element_by_css_selector(all).click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
starbucks_soup_list = soup.select('li.quickResultLstCon')
starbucks_soup_list[0]
```

```python
starbucks_list=[]
for item in starbucks_soup_list :
    name = item.select('strong')[0].text.strip()
    lat = item['data-lat'].strip()
    lng = item['data-long'].strip()
    addr = str(item.select('p.result_details')[0]).split('<br/>')[0].split('>')[1]
    tel = str(item.select('p.result_details')[0]).split('<br/>')[1].split('<')[0]

    starbucks_list.append([name,lat,lng,addr,tel])
starbucks_list
```

```python
columns =['매장명','위도','경도','주소','전화번호']
seoul_starbucks_df = pd.DataFrame(starbucks_list, columns = columns)
seoul_starbucks_df
```

```python
seoul_starbucks_df.to_excel('seoul_starbucks2.xlsx', index=False)
```

```python
dong_names = []

for addr in seoul_starbucks_df['주소']:
    dong = addr.split()[1]
    dong_names.append(dong)

seoul_starbucks_df['시군구명'] = dong_names

seoul_starbucks_df.head()
```

```python
seoul_starbucks_df.to_excel('seoul_starbucks3.xlsx', index=False)
```

```python
seoul_starbucks_sgg_count = seoul_starbucks_df.pivot_table(index='시군구명',
                                                        values='매장명',
                                                        aggfunc='count').rename(columns={'매장명':'스타벅스_매장수'})
seoul_starbucks_sgg_count.head()
```

```python
seoul_sgg = pd.read_excel('sgg_pop.xlsx')
```

```python
seoul_sgg = pd.merge(seoul_sgg, seoul_starbucks_sgg_count, how = 'left', on='시군구명')
seoul_sgg.head()
```

```python
seoul_biz = pd.read_excel('sgg_biz.xlsx')
seoul_biz.head()
```

```python
seoul_sgg = pd.merge(seoul_sgg, seoul_biz, how = 'left', on='시군구명')
seoul_sgg.head()
```

```python
seoul_sgg.to_excel('seoul_sgg_all.xlsx', index=False)
```

```python
import folium
```

```python
seoul_starbucks = pd.read_excel('./seoul_starbucks_list.xlsx')
```

```python
starbucks_map = folium.Map(
    location=[37.57, 126.98],
    zoom_start=12,
    tiles='Stamen Terrain'
)
starbucks_map
```

```python
for idx in seoul_starbucks.index:
    lat = seoul_starbucks.loc[idx,'위도']
    lng = seoul_starbucks.loc[idx,'경도']

    folium.Marker(
        location = [lat, lng]
    ).add_to(starbucks_map)
starbucks_map
```
